```python
stemmer = PorterStemmer()
wnl = WordNetLemmatizer()
```

```python
#Removing any punctuations from the review text and converting all the text to lower case
df['processed_words'] = df['no_stopwords'].apply(lambda x: re.sub('[^a-zA-Z\s]', '', x))
df['processed_words'] = df['processed_words'].apply(lambda x: re.sub('\s+', ' ', x))
```

```python
#Stemmering the data to remove the parts of speech
df['processed_words'] =  df['processed_words'].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))
```

```python
#Lammetizing the data to male plural words to singular
df['processed_words'] =  df['processed_words'].apply(lambda x: ' '.join([wnl.lemmatize(word) for word in word_tokenize(x)]))
```

```python
print(df['processed_words'])
```

- The process of reducing a word to its base form is called lemmatization. Unlike stemming, which only removes the last few characters of a word, lemmatization takes into account the context of the word and converts it to its most meaningful base form. As a result, stemming often results in incorrect meanings and spelling errors.
- A method called stemming is utilized to shorten an inflected word to its stem, as illustrated by the words "programming," "programmer," and "programs," which can all be reduced to the root word stem "program."
- Stemming and lemmatization are methods used by search engines and chatbots to analyze the meaning behind a word. Stemming uses the stem of the word, while lemmatization uses the context in which the word is being used.
- <b>Lemmatization has higher accuracy than stemming. Lemmatization is preferred for context analysis, whereas stemming is recommended when the context is not important.

Utilizes the TfidfVectorizer to convert text data into TF-IDF features, with various preprocessing steps and a maximum of 10,000 features, resulting in a transformed matrix representation for the training text.

```python
word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    stop_words='english',
    ngram_range=(1, 1),
    max_features=10000)
word_vectorizer.fit(all_text)
train_word_features = word_vectorizer.transform(train_text)
```

The provided code defines a character-level TfidfVectorizer that converts text data into TF-IDF features, followed by fitting the vectorizer on the entire text corpus, transforming the training text into a character-level TF-IDF matrix, and finally combining the character and word-level features using hstack to create the train_features matrix.

```python
char_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='char',
    stop_words='english',
    ngram_range=(2, 6),
    max_features=50000)
char_vectorizer.fit(all_text)
train_char_features = char_vectorizer.transform(train_text)

train_features = hstack([train_char_features, train_word_features])
```

```python
#Splitting the data into 80% and 20%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(train_features, target, train_size = 0.8)
```

## K Neighbors Classifier(KNN)

The KNeighborsClassifier is a machine learning algorithm used for classification tasks. It falls under the category of instance-based or lazy learning algorithms. Unlike traditional algorithms, it does not explicitly learn a model during the training phase. Instead, during prediction, it identifies the k-nearest neighbors in the training dataset based on the distance between samples. The number of neighbors, k, is a hyperparameter that determines the number of neighbors considered. Classification is then determined by a majority vote among the neighbors, with the class having the highest count being assigned as the predicted class for the input sample. KNeighborsClassifier is versatile and can be used for both binary and multi-class classification problems. However, the choice of distance metric used for calculating the distances between samples is important and can impact performance. The algorithm stores the training data in memory, allowing for efficient computation during prediction. It is particularly useful for problems with non-linear or complex decision boundaries and is commonly employed in pattern recognition tasks.

```python
from sklearn import neighbors
classifier=neighbors.KNeighborsClassifier()
```

The below provided code creates and evaluates different classification models (logistic regression, decision tree, random forest, and support vector machine) using feature selection and accuracy calculation. It initializes the models and stores them in a dictionary. The code also defines a list of values for the number of top features to select. It then iterates over the models and feature values, creating a pipeline that includes feature selection and the classification model. The pipeline is fitted on the training data, and accuracy scores are calculated for both the training and testing datasets. The accuracy scores are stored in a DataFrame for each model and feature value combination. This allows for the comparison of model performance based on different feature selections. Overall, the code automates the process of evaluating multiple models with different feature selections and provides a means to assess their accuracy.

```python
lr_model = LogisticRegression(solver='liblinear')
dt_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
svc_model = SVC(kernel = 'rbf')

# Storing those models in a dictionary
models = {'Logistic_Regression': lr_model,
          'Decision_Tree_Classifier':dt_model,
          'Random_Forest_Classifier': rf_model,
          'Support_Vector_Machine': svc_model}

# storing the values of k which is the top k-features we use by using CHI2 and Select K best
k_features = [5, 100, 500, 10000]


# Iterating over models and k values and calculating accuracy scores
for model_name, model in models.items():
    globals()[model_name] = pd.DataFrame(index = k_features, columns = ['Train_accuracy', 'Test_accuracy'])
    for k in k_features:
        
        # This is a pipeline of vectorization, selectkbest and a classification model
        piped_model = Pipeline(steps=[("Selecting K best features", SelectKBest(score_func=chi2, k=k)),
                           ("classifier", model)])
        piped_model.fit(X_train, y_train)
        
        globals()[model_name].loc[k,:] = [accuracy_score(piped_model.predict(X_train), y_train),
                                           accuracy_score(piped_model.predict(X_test), y_test)]
```

```python
def plot_graph(df_name):
    
    df = globals()[df_name]
    df = df.sort_index(ascending = False)
    
    k_features_reverse = k_features[::-1]
    sample = ['Top_' + str(k) + '_features' for k in k_features_reverse]

    x_axis = np.arange(len(sample)) * 1.5

    # Multi bar Chart

    plt.figure(figsize = (10, 3))
    plt.bar(x_axis +0.20, df['Train_accuracy'], width=0.2, label = 'Train Accuracy')
    plt.bar(x_axis +0.20*2, df['Test_accuracy'], width=0.2, label = 'Test Accuracy')
    
    
    # Title
    plt.title(df_name, fontweight = "bold")
    plt.xlabel('Top K features')
    plt.ylabel('Accuracy Score')
    

    # Xticks

    plt.xticks(x_axis + 0.4, sample, rotation = 90)

    # Add legend

    plt.legend(loc = 'upper left', bbox_to_anchor=(0, 1.3))

    # Display

    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Select the categorical columns from the DataFrame
categorical_df = df[cat_cols]

# Calculate the chi-square statistic and p-value for each categorical column with the text column
chi2_scores = []
p_values = []
for col in categorical_df.columns:
    contingency_table = pd.crosstab(df[col], processed_text[text_col])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    chi2_scores.append(chi2)
    p_values.append(p_value)

# Create a DataFrame to store the chi-square scores and p-values
chi2_results = pd.DataFrame({'Column': categorical_df.columns, 'Chi2 Score': chi2_scores, 'p-value': p_values})

# Sort the DataFrame by p-value in ascending order
chi2_results.sort_values(by='p-value', inplace=True)

# Create a bar plot to visualize the p-values
plt.figure(figsize=(10, 6))
sns.barplot(data=chi2_results, x='p-value', y='Column', palette='coolwarm')
plt.xlabel('p-value')
plt.ylabel('Categorical Column')
plt.title('p-value of Chi-square Test for Categorical Columns vs. Text Column')
plt.show()
```

The p-value measures the statistical significance of the relationship between each categorical column and the text column. A lower p-value indicates a stronger association between the two variables.

Please note that the chi-square test assumes independence between the categorical variables and the text column. It tests the null hypothesis that there is no association between the variables. However, it's important to interpret the results cautiously and consider the context of your data.

# Long Short Term Memory Network

LSTMs, or Long Short-Term Memory networks, are a type of RNNs that are commonly used in NLP and speech recognition. LSTMs are designed to handle long-term dependencies and overcome the vanishing gradient problem in traditional RNNs. They do this by maintaining a cell state and three gates, which allow them to selectively forget, remember, and update information over time. LSTMs have been successfully applied to a variety of NLP tasks, such as machine translation, speech recognition, and sentiment analysis.

```python
import numpy as np
import pandas as pd
from keras.models import Model
from keras.layers import Input, Embedding, Dense, concatenate
from keras.layers import LSTM, Bidirectional, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

X_num = df[num_cols].values
X_cat = df[cat_cols].values
X_text = processed_text[text_col].values
y = y_df.values

cat_onehot = []
for i in range(len(cat_cols)):
    col_onehot = pd.get_dummies(X_cat[:, i], prefix=cat_cols[i])
    cat_onehot.append(col_onehot.values)
X_cat = np.concatenate(cat_onehot, axis=1)

tokenizer = Tokenizer(num_words=275)
tokenizer.fit_on_texts(X_text)
X_text = tokenizer.texts_to_sequences(X_text)
X_text = pad_sequences(X_text, maxlen=270)

# Convert the target variable to categorical format
y = to_categorical(y)

X_train_num, X_val_num, X_train_cat, X_val_cat, X_train_text, X_val_text, y_train, y_val = train_test_split(
    X_num, X_cat, X_text, y, test_size=0.2, random_state=42)
X_val_num, X_test_num, X_val_cat, X_test_cat, X_val_text, X_test_text, y_val, y_test = train_test_split(
    X_val_num, X_val_cat, X_val_text, y_val, test_size=0.5, random_state=42)

num_inputs = Input(shape=(len(num_cols),))
cat_inputs = Input(shape=(X_cat.shape[1],))
text_inputs = Input(shape=(270,))
embeddings = Embedding(input_dim=10000, output_dim=50, input_length=270)(text_inputs)
lstm = Bidirectional(LSTM(units=64, dropout=0.2, recurrent_dropout=0.2))(embeddings)
x = concatenate([num_inputs, cat_inputs, lstm])
x = Dense(units=64, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(units=32, activation='relu')(x)
x = Dropout(0.2)(x)
outputs = Dense(units=len(np.unique(y_df)), activation='softmax')(x)  # Set units to the number of classes
model = Model(inputs=[num_inputs, cat_inputs, text_inputs], outputs=outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit([X_train_num, X_train_cat, X_train_text], y_train, epochs=1, batch_size=32, validation_data=([X_val_num, X_val_cat, X_val_text], y_val))

loss, accuracy = model.evaluate([X_test_num, X_test_cat, X_test_text], y_test)
print('Test accuracy:', accuracy)
```

The LSTM model achieved a test accuracy of 89%, indicating its reasonable performance in predicting the target variable

# Convolution Neural Network
CNNs, or Convolutional Neural Networks, are typically used in computer vision tasks but can also be used in NLP for feature extraction from text. CNNs can capture local dependencies between words and handle variable-length input sequences by padding inputs with zeros or truncating them to a fixed length. CNNs have been used for text classification, sentiment analysis, and question answering and have been combined with other neural networks to improve performance on NLP tasks. Overall, CNNs offer a powerful and flexible approach to NLP tasks.

```python
import numpy as np
import pandas as pd
from keras.models import Model
from keras.layers import Input, Embedding, Dense, concatenate
from keras.layers import Conv1D, GlobalMaxPooling1D, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

X_num = df[num_cols].values
X_cat = df[cat_cols].values
X_text = processed_text[text_col].values
cat_onehot = []
for i in range(len(cat_cols)):
    col_onehot = pd.get_dummies(X_cat[:, i], prefix=cat_cols[i])
    cat_onehot.append(col_onehot.values)
X_cat = np.concatenate(cat_onehot, axis=1)

tokenizer = Tokenizer(num_words=275)
tokenizer.fit_on_texts(X_text)
X_text = tokenizer.texts_to_sequences(X_text)
X_text = pad_sequences(X_text, maxlen=270)

X_train_num, X_val_num, X_train_cat, X_val_cat, X_train_text, X_val_text, y_train, y_val = train_test_split(
    X_num, X_cat, X_text, y_df, test_size=0.2, random_state=42)
X_val_num, X_test_num, X_val_cat, X_test_cat, X_val_text, X_test_text, y_val, y_test = train_test_split(
    X_val_num, X_val_cat, X_val_text, y_val, test_size=0.5, random_state=42)

num_inputs = Input(shape=(len(num_cols),))
cat_inputs = Input(shape=(X_cat.shape[1],))
text_inputs = Input(shape=(270,))

embeddings = Embedding(input_dim=10000, output_dim=50, input_length=270)(text_inputs)
conv1 = Conv1D(filters=64, kernel_size=3, activation='relu')(embeddings)
pool1 = GlobalMaxPooling1D()(conv1)

x = concatenate([num_inputs, cat_inputs, pool1])
x = Dense(units=128, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(units=64, activation='relu')(x)
x = Dropout(0.5)(x)
outputs = Dense(units=2, activation='softmax')(x)

model = Model(inputs=[num_inputs, cat_inputs, text_inputs], outputs=outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit([X_train_num, X_train_cat, X_train_text], to_categorical(y_train), epochs=1, batch_size=32,
          validation_data=([X_val_num, X_val_cat, X_val_text], to_categorical(y_val)))

loss, accuracy = model.evaluate([X_test_num, X_test_cat, X_test_text], to_categorical(y_test))
print('Test accuracy:', accuracy)
```

The CNN model achieved a test accuracy of 89%, indicating its reasonable performance in predicting the target variable(fake review detection). Both LSTM and CNN models achieved similar high accuracies, indicating that both architectures are suitable for the given task
